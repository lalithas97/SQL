19. Check if date is weekend.
```sql
select  to_char(ORDER_DATE,'DY'),
        case
            when to_char(order_date,'DY') in ('SAT','SUN')
            then 'Weekend'
            else 'Weekday'
       end
from orders
```
OR
```sql
select to_char(order_date,'DAY'),
        case
          when trim(to_char(order_date,'DAY')) in ('SATURDAY','SUNDAY')
          then 'weekend'
          else 'weekday'
        end
from orders
```

We have to use trim because TO_CHAR(order_date, 'DAY')

returns fixed-length padded text.
Ex: 'SATURDAY   '

So 'SATURDAY   '  ≠  'SATURDAY'
